# Creating a function to plot our model's prediction -> plot_decision_boundary()

* it will take in trained model's feature (X) and labels (y)
* create meshgrid of different values of x 
* make prediction on the different value of meshgrid 
* plot the prediction as well as the line between different class of the input

```python
def plot_decision_boundary(model, X,y):
    '''
      Plots the decision boundary created by a model predicting on X.
      This function has been adapted from two phenomenal resources :
           1. CS231n - https://cs231n.github.io/neural-networks-case-study/
           2. Made with ML basics - https://github.com/GokuMohandas/MadeWithML/blob/main/notebooks/08_Neural_Networks.ipynb
    '''
    # define the axis boundaries and create a meshgrid
    x_min , x_max=X[:,0].min() - 0.1 ,X[:,0].max() +0.1
    y_min ,y_max =X[:,1].min()-0.1 ,X[:,1].max()+0.1
    xx,yy=np.meshgrid(np.linspace(x_min,x_max,100),
    np.linspace(y_min,y_max,100))                  
    # create X- values (we 're going to make predictions on these)
    x_in = np.c_[xx.ravel(),yy.ravel()]
    # make predictions 
    y_pred = model.predict(x_in)
#     check for multiclass 
    if len(y_pred[0])>1:
        print("doing multiclass classification")
#       we have to reshape our prediction to get them ready for classification
        y_pred=np.argmax(y_pred,axis =1).reshape(xx.shape)
    
    else :
        print("doing binary classification")
        y_pred=np.round(y_pred).reshape(xx.shape)
#  plot the decision boundary 
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha =0.7)
    plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap =plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

# Using callbacks to find the best Learning Rate for our model's
**To find the ideal learning rate for our model where the loss decrease the most , we are going to create a callback function to fetch us that learning rate**
We can think of a callback as an extra piece of functionality which determine the ideal learning rate for our model **while its training**


```python
# creating the model same as model_8
# set random seed
tf.random.set_seed(42)
# design the model 
model_9 = tf.keras.Sequential([
    tf.keras.layers.Dense(4,activation ="relu"),
    tf.keras.layers.Dense(4,activation ="relu"),
    tf.keras.layers.Dense(1,activation ="sigmoid")
])
model_9.compile(loss="binary_crossentropy",optimizer = "Adam",metrics =["accuracy"])
# creating a callback 
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))# this lambda function just tells the model that after every epoch,update the learning rate according to the given formula 
history_9 = model_9.fit(X,y,epochs =100 ,callbacks =[lr_scheduler]) 
```

```python
pd.DataFrame(history_9.history).plot(figsize = (10,7))
```

# Finding the ideal learning rate for our multi- class classifier model....

```python
tf.random.set_seed(42)

model_13 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation = 'relu'),
    tf.keras.layers.Dense(4,activation ='relu'),
    tf.keras.layers.Dense(10,activation =tf.keras.activations.softmax)
])
# create a learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 *10 **(epoch/20))
# compile the model
model_13.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),optimizer = tf.keras.optimizers.Adam(),metrics =['accuracy'])
```

```python
find_lr_history =model_13.fit(train_data_norm,train_labels,epochs = 40,validation_data =(test_data_norm,test_labels),callbacks =[lr_scheduler])
```

```python
# plotting the learning rate vs loss curve 
lrs = 1e-3*10**(tf.range(40)/20)
plt.semilogx(lrs,find_lr_history.history['loss'])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
```

### from the graph above it can be concluded that the ideal learning rate value for the model is somewhere around 1e-3 which is the default value for the Adam Optimizer....

```python
def plot_confusion_matrix(y_test,y_pred,classes=None,figsize=(10,10),text_size=20):
    cm = confusion_matrix(y_test, tf.round(y_pred))
    cm_norm = cm.astype('float')/cm.sum(axis=1)
#     Label_dict = {0:'T-Shirt/Top',1:'Trouser',2:'Pull Over',3:'Dress',4:'Coat',5:'Sandal',6:'Shirt',7:'Sneaker',8:'Bag',9:'Ankle Boot'}
    n_classes = cm.shape[0]

    fig, ax = plt.subplots(figsize = figsize)
    # creating a matrix plot
    cax = ax.matshow(cm,cmap = plt.cm.Blues)
    fig.colorbar(cax)

    if classes:
        labels = classes
    else :
        labels = np.arange(cm.shape[0])
        
    ax.set(title = "Confusion Matrix", 
           xlabel ="Predicted Labels",
            ylabel ="True Labels", 
            xticks = np.arange(n_classes),
            yticks =np.arange(n_classes),
          xticklabels =labels,
          yticklabels = labels)
    
    ax.yaxis.label.set_size(text_size)
    ax.xaxis.label.set_size(text_size)
    ax.title.set_size(text_size)
    
    ax.xaxis.set_label_position("bottom")
    ax.xaxis.tick_bottom()
    
    threshold = (cm.max()+ cm.min())/2
    
    # plotting the text on each cell using itertool
    
    for i , j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
        plt.text(j,i,f"{cm[i,j]} ({cm_norm[i,j]*100:.2f}%", 
                horizontalalignment ="center", 
                color ="white" if cm[i,j]>threshold else "black",size =15)

        
```

Often when working with image and other form of visual data its a good idea to visualize the data to develop a better understanding of the data and the inputs and the outputs of our model ,,,,,

Creating a little function 
* that plot a random image 
* predict on the generated image 
* label the plot with predicted and true label 


```python
import random
def lilfunc(train_data, train_labels,model,classes):
    '''A little function that generates a random image and predicts its label and then compare it with the true label'''
    number= random.randint(0,len(train_data))
    plt.imshow(train_data[number])
    true_label = train_labels[number]
    predicted_label = model.predict(train_data[number])
    out = predicted_label.argmax(axis =1)
    out = int(out)
    print("Actual Label - ",classes[true_label])
    print("Predicted Label - ",classes[out])
    
    if out == true_label:
        color ='green'
    else :
        color ='red'
    plt.xlabel("Predicted label = {} ,  Confidence = {:2.0f}% ,  true label = {}".format(classes[out],100*tf.reduce_max(predicted_label),classes[true_label]),color =color)
```
